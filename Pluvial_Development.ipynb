### Precipitation for Hydrologic Modeling 

Pluvial flood risk at an area of concern is linked to the rainfall received.  The contributors to flood risk are, rainfall duration, temporal distribution of rainfall and depth (total rainfall received). When area of interest is large, the areal coverage of the rainfall and watershed drainage become contributors as well.  Flood risk due to pluvial source comes in two stages: by direct rainfall and that enters the area of interest through fluvial means – drainage of the upstream watershed.  To simplify the risk formulation, pluvial flood hazard is assumed to be generated by direct rainfall only.  Waters that enter the area of concern through fluvial means – drainage of upstream watershed – is not considered for this study.

National Oceanic and Atmospheric agency (NOAA) has analyzed the rain gauge records nationally and published precipitation frequency estimates and 90% confidence intervals, for the United States and Affiliated Territories, in [__Atlas 14__](https://hdsc.nws.noaa.gov/hdsc/pfds/pfds_map_cont.html).  Precipitation frequency estimates for 19 durations ranging from 5 minutes to 60 days, and 10 return periods (1 through 1000 year)  are published in NOAA’s  Precipitation Frequency Data Server for all States except for six (WA, OR, ID, MT, WY, and TX.).  

For storms of 6-, 12-, 24-, and 96-hour duration, Atlas 14 provides 36 plausible temporal distributions.  Using the occurrence of the peak frequency intensity, four temporal distributions - first through forth quartile – are published. Based on the percentage of occurrence (varying between 10% and 90% of storms used for the Atlas 14 study) further nine (decile) distributions are developed.

Atlas 14 precipitation estimates are point rainfall estimates and are not directly applicable to an area.  To extend these rainfall estimate to cover an area, areal reduction factors needs to be applied.  These factors are established by U.S Weather Bureau in Technical Paper 90 for watersheds with drainage areas less than 400 sq. mi. and for storms of duration of 24 hours or less.  For larger watersheds, a site-specific analysis may be appropriate to model pluvial flooding.
For this study, pluvial flood risk is estimated using rainfalls of durations 6 hours, 12 hours, 24 hours, and 96-hours.  Rainfall frequency is taken from the Atlas 14 study results.  Uncertainty in rainfall distribution is defined by the 36 temporal distributions published in Atlas 14.
